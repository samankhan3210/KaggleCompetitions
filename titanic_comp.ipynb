```python
import pandas as pd
train_df=pd.read_csv("/content/drive/MyDrive/kaggle/train.csv")
df = pd.read_csv("/content/drive/MyDrive/kaggle/train.csv")
df2 = pd.read_csv("/content/drive/MyDrive/kaggle/test.csv")
```

```python
df.head()
```

```python
def familygroupcreation(FG): # Creating a user-defined function for creating Family buckets
    FamilyGroup = [] # Array declartion
    if FG == 1 : 
        FamilyGroup = "0" 
    if FG > 1 and FG < 5 : 
        FamilyGroup = "1"       
    if FG >= 5 :
        FamilyGroup = "2"

    return FamilyGroup

df['FamilyCount'] = 1 + df['SibSp'] + df['Parch'] 
Family_Group = [familygroupcreation(FG) for FG in df.FamilyCount] # Function call
df.insert(3, "FamilyGroup", Family_Group) 
df2['FamilyCount'] = 1 + df2['SibSp'] + df2['Parch'] 
Family_Group = [familygroupcreation(FG) for FG in df2.FamilyCount] # Function call
df2.insert(3, "FamilyGroup", Family_Group) 
```

```python
print(df.isnull().sum())
print(df2.isnull().sum())
```

```python
import numpy as np

df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
pid = df2['PassengerId'].tolist()
df2.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna('S')
df2["Age"] = df2["Age"].fillna(df2["Age"].mean())
df2["Fare"] = df2["Fare"].fillna(df2["Fare"].mean())
Q1 = df.Fare.quantile(0.25)
Q2 = df.Fare.quantile(0.50)
Q3 = df.Fare.quantile(0.75)

print("\n'Very Low Fare' Range:", 0, "-", Q1.round(2))
print("\n'Low Fare' Range:", Q1.round(2), "-", Q2.round(2))
print("\n'Medium Fare' Range:", Q2.round(2), "-", Q3.round(2))
print("\n'High Fare' Range:", Q3, "-", round(max(df.Fare), 2), '\n\n')

#Creating fare groups
df.insert(10, 'FareGroup', np.nan)

df.loc[(df.Fare <= Q1), 'FareGroup'] = 'VeryLowFare'
df.loc[(df.Fare > Q1) & (df.Fare <= Q2), 'FareGroup'] = 'LowFare'
df.loc[(df.Fare > Q2) & (df.Fare <= Q3), 'FareGroup'] = 'MediumFare'
df.loc[(df.Fare > Q3), 'FareGroup'] = 'HighFare'
print(df.isnull().sum())
print(df2.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

col1 = ['Sex','Embarked', 'FareGroup']

for j in col1:
  l = LabelEncoder()
  df[j] = l.fit_transform(df[j])
  df2[j] = l.fit_transform(df2[j])

df.head()
```

```python

age = []
for i in df['Age']:
  if i <= 18:
    age.append(0)
  elif i > 18 and i <= 45:
    age.append(1)
  else:
    age.append(2)

age2 = []
for i in df2['Age']:
  if i <= 18:
    age2.append(0)
  elif i > 18 and i <= 22:
    age2.append(1)
  else:
    age2.append(2)

df['Age'], df2['Age'] = age, age2
```

```python
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from xgboost import XGBRFClassifier

X_tr, X_t, y_tr, y_t = train_test_split(df.drop(['Survived'], axis=1).values, df['Survived'].values, test_size=0.2, random_state=42)
# X_tr, y_tr = df.drop(['Survived'], axis=1).values, df['Survived'].values
# X_t = df2.values 
model = LGBMClassifier()
model.fit(X_tr,y_tr)
y_p = model.predict(X_t)
print("LGBM : ", accuracy_score(y_t, y_p))

model=AdaBoostClassifier()
model.fit(X_tr,y_tr)
y_p = model.predict(X_t)
print("AdaBoostClassifier : ", accuracy_score(y_t, y_p))

model=RandomForestClassifier()
model.fit(X_tr,y_tr)
y_p = model.predict(X_t)
print("RandomForestClassifier : ", accuracy_score(y_t, y_p))

model=DecisionTreeClassifier()
model.fit(X_tr,y_tr)
y_p = model.predict(X_t)
print("DecisionTreeClassifier : ", accuracy_score(y_t, y_p))

model_xgbrf = XGBRFClassifier(n_estimators=300, subsample=0.9, colsample_bynode=0.2)
model_xgbrf.fit(X_tr,y_tr)
y_xgbrf_pred = model_xgbrf.predict(X_t)
print("XGBRF : ", accuracy_score(y_t, y_p))

model_xgb = XGBClassifier()
model_xgb.fit(X_tr,y_tr)
print("XGB : ", accuracy_score(y_t, model_xgb.predict(X_t)))
```
